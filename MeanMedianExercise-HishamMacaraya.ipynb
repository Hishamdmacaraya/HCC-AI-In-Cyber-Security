Here's some code that will generate some random e-commerce data; just an array of total amount spent per transaction. Select the code block, and hit "play" to execute it:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 20.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

Now, find the mean and median of this data. In the code block below, write your code, and see if your result makes sense:

```python
# Calculate the mean
mean_income = np.mean(incomes)

# Calculate the median
median_income = np.median(incomes)

# Print the results
print(f"Mean Income: {mean_income}")
print(f"Median Income: {median_income}")
```

Try playing with the code above to generate different distributions of data, or add outliers to it to see their effect.

```python
# Generate different distributions
incomes_normal = np.random.normal(100.0, 20.0, 10000)
incomes_uniform = np.random.uniform(50.0, 150.0, 10000)
incomes_exponential = np.random.exponential(50.0, 10000)

# Add outliers to each distribution
outliers = [1000, 1200, 1500]
incomes_normal_with_outliers = np.append(incomes_normal, outliers)
incomes_uniform_with_outliers = np.append(incomes_uniform, outliers)
incomes_exponential_with_outliers = np.append(incomes_exponential, outliers)

# Function to calculate and print mean and median
def calculate_and_print_stats(data, label):
    mean = np.mean(data)
    median = np.median(data)
    print(f"{label} - Mean: {mean}, Median: {median}")

# Calculate and print stats for each distribution
calculate_and_print_stats(incomes_normal, "Normal Distribution (original)")
calculate_and_print_stats(incomes_normal_with_outliers, "Normal Distribution (with outliers)")
calculate_and_print_stats(incomes_uniform, "Uniform Distribution (original)")
calculate_and_print_stats(incomes_uniform_with_outliers, "Uniform Distribution (with outliers)")
calculate_and_print_stats(incomes_exponential, "Exponential Distribution (original)")
calculate_and_print_stats(incomes_exponential_with_outliers, "Exponential Distribution (with outliers)")

# Plot histograms for each distribution
plt.figure(figsize=(15, 10))

plt.subplot(3, 2, 1)
plt.hist(incomes_normal, 50, alpha=0.5, label='Normal (original)')
plt.legend(loc='upper right')

plt.subplot(3, 2, 2)
plt.hist(incomes_normal_with_outliers, 50, alpha=0.5, label='Normal (with outliers)')
plt.legend(loc='upper right')

plt.subplot(3, 2, 3)
plt.hist(incomes_uniform, 50, alpha=0.5, label='Uniform (original)')
plt.legend(loc='upper right')

plt.subplot(3, 2, 4)
plt.hist(incomes_uniform_with_outliers, 50, alpha=0.5, label='Uniform (with outliers)')
plt.legend(loc='upper right')

plt.subplot(3, 2, 5)
plt.hist(incomes_exponential, 50, alpha=0.5, label='Exponential (original)')
plt.legend(loc='upper right')

plt.subplot(3, 2, 6)
plt.hist(incomes_exponential_with_outliers, 50, alpha=0.5, label='Exponential (with outliers)')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
