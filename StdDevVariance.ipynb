# Standard Deviation and Variance

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 50.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std()
```

```python
incomes.var()
```

## Activity

Experiment with different parameters on the normal function, and see what effect it has on the shape of the distribution. How does that new shape relate to the standard deviation and variance?

```python
# Function to plot normal distributions with different parameters
def plot_normal_distributions(means, std_devs, num_samples=1000):
    # Generate a range of x values from -10 to 10
    x = np.linspace(-10, 10, 1000)
    
    # Create a figure with a specified size
    plt.figure(figsize=(12, 6))
    
    # Loop through each mean and standard deviation pair
    for mean, std_dev in zip(means, std_devs):
        # Calculate the probability density function (PDF) of the normal distribution
        y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
        
        # Plot the normal distribution
        plt.plot(x, y, label=f'Mean: {mean}, Std Dev: {std_dev}')
    
    # Label the x-axis
    plt.xlabel('Value')
    
    # Label the y-axis
    plt.ylabel('Probability Density')
    
    # Add a title to the plot
    plt.title('Normal Distributions with Different Parameters')
    
    # Add a legend to the plot
    plt.legend()
    
    # Display the plot
    plt.show()

# Experiment with different means and standard deviations
means = [0, 0, 0, 2]  # List of means for the normal distributions
std_devs = [1, 2, 0.5, 1]  # List of standard deviations for the normal distributions

# Call the function to plot the normal distributions
plot_normal_distributions(means, std_devs)
```

### Explaination:

To experiment with different parameters on the normal function and observe their effects on the shape of the distribution, I used the numpy library to generate normal distributions with varying means and standard deviations. I then visualized these distributions using matplotlib to analyze how changes in parameters affect the shape, standard deviation, and variance. By defining a function that takes lists of means and standard deviations, I generated normal distributions and plotted them. I experimented with different means and standard deviations to see their impact on the distribution's shape. I found that changing the mean shifts the distribution along the x-axis, moving the center of the distribution. Adjusting the standard deviation controls the spread of the distribution: a larger standard deviation results in a wider and flatter distribution, while a smaller standard deviation results in a narrower and taller distribution. The variance, being the square of the standard deviation, also affects the spread in the same way. This exercise helped me understand the relationship between the parameters, standard deviation, and variance, and how they influence the shape of the normal distribution. The new shape of the distribution directly relates to the standard deviation and variance, as these parameters determine the width and height of the curve, illustrating the degree of dispersion around the mean.
